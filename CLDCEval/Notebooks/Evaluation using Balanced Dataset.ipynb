```python
import _pickle as cPickle
import matplotlib.pyplot as plt
from collections import Counter
%matplotlib inline
```

## CNN Results:

```python
results_path = "/Users/MeryemMhamdi/Downloads/CNNModels_RCV-BAL/"

path_multi = "en,de,fr,it_joint_emb_ferreira_2016_reg-l1_mu-1e-9_epochs-50_results.p"
path_en_mono = "english_mono_results.p"
path_de_mono = "german_mono_results.p"
path_fr_mono = "french_mono_results.p"
path_it_mono = "italian_mono_results.p"

with open(results_path+path_multi, "rb") as file:
    results_multi = cPickle.load(file, encoding='latin1')
    
with open(results_path+path_en_mono, "rb") as file:
    results_en_mono = cPickle.load(file, encoding='latin1')
    
with open(results_path+path_de_mono, "rb") as file:
    results_de_mono = cPickle.load(file, encoding='latin1')
    
with open(results_path+path_fr_mono, "rb") as file:
    results_fr_mono = cPickle.load(file, encoding='latin1')
    
with open(results_path+path_it_mono, "rb") as file:
    results_it_mono = cPickle.load(file, encoding='latin1')
```

```python
epoch = 0
```

```python
f1_multi_en = []
f1_multi_de = []
f1_multi_fr = []
f1_multi_it = []
f1_mono_en = []
f1_mono_de = []
f1_mono_fr = []
f1_mono_it = []
for i in range(0,20):
    f1_multi_en.append(results_multi["test_metrics_en"][i]["f1_macro"])
    f1_multi_de.append(results_multi["test_metrics_de"][i]["f1_macro"])
    f1_multi_fr.append(results_multi["test_metrics_fr"][i]["f1_macro"])
    f1_multi_it.append(results_multi["test_metrics_it"][i]["f1_macro"])
    
    f1_mono_en.append(results_en_mono["test_metrics_en"][i]["f1_macro"])
    f1_mono_de.append(results_de_mono["test_metrics_de"][i]["f1_macro"])
    f1_mono_fr.append(results_fr_mono["test_metrics_fr"][i]["f1_macro"])
    f1_mono_it.append(results_it_mono["test_metrics_it"][i]["f1_macro"])
```

```python
plt.figure(figsize=[8,6])
plt.plot(f1_multi_en,'r',linewidth=3.0)
plt.plot(f1_mono_en,'g',linewidth=3.0)
plt.legend(['MULTI English', 'Mono English'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('F1 Score',fontsize=16)
plt.title('F1 Score Evolution per epoch',fontsize=16)
```
