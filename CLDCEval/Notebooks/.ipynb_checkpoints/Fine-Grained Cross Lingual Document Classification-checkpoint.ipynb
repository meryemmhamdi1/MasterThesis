### Multilingual Embeddings:

```python
with open(dir_path+"en,de,fr,it_expert_dict_dim_red_en_de_fr_it.txt_results.p", "rb") as file:
    results_multi = pkl.load(file, encoding="latin1")
```

```python
f1_scores_en_multi = []
for i in range(0,len(results_multi["test_metrics_en"])):
    f1_scores_en_multi.append(results_multi["test_metrics_en"][i]["f1_macro"])
print("English f1 score:", f1_scores_en_multi[j])

f1_scores_fr_multi = []
for i in range(0,len(results_multi["test_metrics_fr"])):
    f1_scores_fr_multi.append(results_multi["test_metrics_fr"][i]["f1_macro"])
print("French f1 score:", f1_scores_fr_multi[j])

f1_scores_de_multi = []
for i in range(0,len(results_multi["test_metrics_de"])):
    f1_scores_de_multi.append(results_multi["test_metrics_de"][i]["f1_macro"])
print("German f1 score:", f1_scores_de_multi[j])

f1_scores_it_multi = []
for i in range(0,len(results_multi["test_metrics_it"])):
    f1_scores_it_multi.append(results_multi["test_metrics_it"][i]["f1_macro"])
print("Italian f1 score:", f1_scores_it_multi[j])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=[8,6])
x = list(range(1, len(f1_scores_en)+1))
xi = [i for i in range(1, len(f1_scores_en)+1)]
plt.plot(xi, f1_scores_en,'g',linewidth=3.0)
plt.plot(xi, f1_scores_en_multi,'r',linewidth=3.0)
plt.legend(['Mono English', 'MULT English'],fontsize=18)
plt.xlabel('Epochs ',fontsize=16)
plt.ylabel('F1 Score',fontsize=16)
plt.xticks(xi, x)
plt.title('F1 Score Evolution per epoch',fontsize=16)
```
