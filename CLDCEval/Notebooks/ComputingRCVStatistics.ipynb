```python
import cPickle as pkl
dir_path_en = "Datasets/RCV/MultiLabels/english/"
with open(dir_path_en+ "y_train_single_55_label.p", "rb") as yfile:
    y_train_en = pkl.load(yfile)

with open(dir_path_en+ "y_dev_single_55_label.p", "rb") as yfile:
    y_dev_en = pkl.load(yfile)
    
with open(dir_path_en+ "y_test_single_55_label.p", "rb") as yfile:
    y_test_en = pkl.load(yfile)
```

```python
dir_path_fr = "Datasets/RCV/MultiLabels/french/"
print("Loading y_train")
with open(dir_path_fr+ "y_train_single_55_label.p", "rb") as yfile:
    y_train_fr = pkl.load(yfile)

print("Loading y_dev")
with open(dir_path_fr+ "y_dev_single_55_label.p", "rb") as yfile:
    y_dev_fr = pkl.load(yfile)
    
print("Loading y_test")
with open(dir_path_fr+ "y_test_single_55_label.p", "rb") as yfile:
    y_test_fr = pkl.load(yfile)
```

```python
dir_path_de = "Datasets/RCV/MultiLabels/german/"
print("Loading y_train")
with open(dir_path_de+ "y_train_single_55_label.p", "rb") as yfile:
    y_train_de = pkl.load(yfile)

print("Loading y_dev")
with open(dir_path_de+ "y_dev_single_55_label.p", "rb") as yfile:
    y_dev_de = pkl.load(yfile)
    

print("Loading y_test")    
with open(dir_path_de+ "y_test_single_55_label.p", "rb") as yfile:
    y_test_de = pkl.load(yfile)
```

```python
dir_path_it = "Datasets/RCV/MultiLabels/italian/"
print("Loading y_train")
with open(dir_path_it+ "y_train_single_55_label.p", "rb") as yfile:
    y_train_it = pkl.load(yfile)

print("Loading y_dev")
with open(dir_path_it+ "y_dev_single_55_label.p", "rb") as yfile:
    y_dev_it = pkl.load(yfile)
    
print("Loading y_test")
with open(dir_path_it+ "y_test_single_55_label.p", "rb") as yfile:
    y_test_it = pkl.load(yfile)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


N = 4
train = (len(y_train_en), len(y_train_de), len(y_train_fr), len(y_train_it))
dev = (len(y_dev_en), len(y_dev_de), len(y_dev_fr), len(y_dev_it))
test = (len(y_test_en), len(y_test_de), len(y_test_fr), len(y_test_it))
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, train, width)
p2 = plt.bar(ind, dev, width, bottom=train)
p3 = plt.bar(ind, test, width, bottom=dev)

plt.ylabel('Scores')
plt.title('Documents Distribution across languages')
plt.xticks(ind, ('English', 'German', 'French', 'Italian'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0], p3[0]), ('Train', 'Dev', 'Test'))

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
lMeans = (20, 2, 4, 0, 5)
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, menMeans, width)
p2 = plt.bar(ind, womenMeans, width, bottom=menMeans)
p3 = plt.bar(ind, lMeans, width)

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0],  p3[0]), ('Men', 'Women', 'l'))

plt.show()
```
